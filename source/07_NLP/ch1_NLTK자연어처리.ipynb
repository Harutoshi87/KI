````python
# 오스틴 소설 엠마 내용 데이터 셋
emma = nltk.corpus.gutenberg.raw('austen-emma.txt')
print(emma[:30])
````

````python
type(emma), len(emma)
````

````python
# sent_tokenize() : 문장 단위로 쪼갠 list 반환
from nltk.tokenize import sent_tokenize
sent_tokens = sent_tokenize(emma)
print('문장수 :', len(sent_tokens))
print("%r" % sent_tokens[0])
# %r → "representation"
# repr()의 줄임말
# 개발자용 출력, 문자열을 코드에서 보는 그대로(', \n 등 포함) 보여줌

# %s → "string"
# str()의 줄임말
# 사람용 출력, 그냥 읽기 좋은 형태로 보여줌
# 즉, r = repr, s = str
````

````python
type(sent_tokens)
````

````python
# word_tokenize() : 단어 단위로 쪼갠 list 반환
from nltk.tokenize import word_tokenize
print(word_tokenize(sent_tokens[0]))
````

````python
# RegexpTokenizer 클래스 : 토큰화할 때, 정규표현식
from nltk.tokenize import RegexpTokenizer
ret = RegexpTokenizer('\d+') # [0-9] : \d
digits = ret.tokenize(emma)
print(digits)
ret1 = RegexpTokenizer('\w+') # [a-zA-Z0-9_] : \w
words = ret1.tokenize(sent_tokens[0])
print(words)
ret1 = RegexpTokenizer('[a-zA-Z]+') #  해당 정규표현식에 맞는 단어만 추출
words = ret1.tokenize(sent_tokens[0])
print(words)
````

# 2. 형태소(의미가 있는 가장 작은 단위) 분석
cf. 자연어 처리의 기본은 형태소 분석과 품사태깅 - 어간추출(Stemming), 품사태깅(PosTagging)

````python
words = ['sending', 'cooking', 'files', 'lives', 'crying', 'dying']
# 어간 추출1 : PorterStemmer : Martin Porter라는 사람이 1980년에 만든 영어 어간 추출(Stemming) 알고리즘
from nltk.stem import PorterStemmer, LancasterStemmer, RegexpStemmer
pst = PorterStemmer()
# pst.stem(words[0], words[1])
[pst.stem(word) for word in words]
````

````python
# 어간 추출2 : LancasterStemmer 
lst = LancasterStemmer()
[lst.stem(word) for word in words]
````

````python
# 어간 추출3 : RegexpStemmer
rst = RegexpStemmer('ing') # 기본 옵션 replace=""이므로 'ing'를 지움
# 필요하면 r'ing$'처럼 $를 써서 단어 끝에만 적용하도록 더 엄밀하게 쓸 수도 있음
[rst.stem(word) for word in words]
````

# 3. 품사 태깅

````python
# print('품사태깅할 내용 :', word_tokenize(sent_tokens[0]))

from nltk.tag import pos_tag
tagged_list = pos_tag(word_tokenize(sent_tokens[0]))
print('품사태깅 결과 :', tagged_list)
````

````python
# 명사만 추출 : NN, NNS, NNP, NNPS
# 태그	의미
# NN	단수 일반 명사 (common noun, singular)
# 예: dog, book
# NNS	복수 일반 명사 (common noun, plural)
# 예: dogs, books
# NNP	단수 고유 명사 (proper noun, singular)
# 예: John, London
# NNPS	복수 고유 명사 (proper noun, plural)
# 예: The Smiths, Americas
nouns_list = []
for word, tag in tagged_list:
    #if (tag=='NN') | (tag=='NNS') | (tag=='NNP') |(tag=='NNPS') :
    #if tag in ['NN', 'NNS', 'NNP', 'NNPS'] :
    if tag.find('NN') != -1:
        nouns_list.append(word)
print(nouns_list)
````

````python
nouns_list = [word for word, tag in tagged_list if tag.find('NN') != -1]
print(nouns_list)
````

## Quiz. emma 소설안에서
1. 특수문자가 들어가지 않은 3글자 이상의 단어만 추출해서 품사태깅을 하시오(RegexpTokenizer).

2. "Emma" 단어가 몇번 등장했는지, 품사태깅이 어떤 품사들로 되어 있는지 모두 출력하시오

3. 내가 원하는 품사(명사:NN, NNS, NNP, NNPS)만 뽑아 등장하는 명사의 종류 갯수 출력하시오.

````python
# 1. 특수문자가 들어가지 않은 3글자 이상의 단어만 추출해서 품사태깅을 하시오(RegexpTokenizer)
ret = RegexpTokenizer(r'\b[a-zA-Z]{3,}\b')  # 3글자 이상, 알파벳만 
# \b : 
# 단어 시작/끝 경계 공백, 문장부호, 문자열 시작/끝 등과 맞닿은 위치
# 단어 단위로 정확히 뽑고 싶을 때 필요
words = ret.tokenize(emma)
tagged_words = pos_tag(words)
tagged_words
````

````python
len(tagged_words)
````

````python
type(words)
````

````python
# 2. "Emma" 단어가 몇번 등장했는지, 품사태깅이 어떤 품사들로 되어 있는지 모두 출력하시오
emma_cnt = words.count('Emma')
emma_cnt
````

````python
pos_set = set([tag for word, tag in tagged_words])
print(pos_set, end=' ')
````

````python
# 3. 내가 원하는 품사(명사:NN, NNS, NNP, NNPS)만 뽑아 등장하는 명사의 종류 갯수 출력하시오
nn_tag = {'NN'}
nns_tag = {'NNS'}
nnp_tag = {'NNP'}
nnps_tag = {'NNPS'}
nn = [word for word, tag in tagged_words if tag in nn_tag]
nns = [word for word, tag in tagged_words if tag in nns_tag]
nnp = [word for word, tag in tagged_words if tag in nnp_tag]
nnps = [word for word, tag in tagged_words if tag in nnps_tag]
nn_count = len(set(nn))
nns_count = len(set(nns))
nnp_count = len(set(nnp))
nnps_count = len(set(nnps))
print('일반명사(단수) :', nn_count, '일반명사(복수) :', nns_count, '단수고유명사 :', nnp_count, '복수고유명사 :', nnps_count, sep='\n')
````
