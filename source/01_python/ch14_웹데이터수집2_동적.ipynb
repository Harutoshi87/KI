```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
driver.get('https://www.python.org')
elem = driver.find_element(By.NAME, 'q')
elem.clear()
elem.send_keys('pycon')
```

```python
from selenium.webdriver.common.keys import Keys
elem.send_keys(Keys.RETURN) # Enter를 입력
```

```python
btn_elem = driver.find_element(By.CSS_SELECTOR, 'button#submit') # 버튼 클릭하여 submit
btn_elem.click()
```

```python
result_list = driver.find_elements(By.CSS_SELECTOR, 'li > h3 > a')
for r in result_list:
    title = r.text
    link = r.get_attribute('href')
    print('{} - {}'.format(title, link))
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser') # driver.page_source < vgl. > response / response.text
result_list = soup.select('li > h3 > a')
for r in result_list:
    print('{} - {}'.format(r.text, r.attrs.get('href')))
```

```python
from urllib.parse import urlparse # url을 분석하여 쪼갬
current_url = driver.current_url # 셀레니움을 통해 접근한 url
print(current_url)
parse_url = urlparse(current_url)
print(parse_url)
domain = f'{parse_url.scheme}://{parse_url.netloc}'
print(domain)
```

```python
result_list = soup.select('li > h3 > a')
for r in result_list:
    print('{} - {}'.format(r.text, domain + r.attrs.get('href')))
```

```python
driver.close() # 브라우저 닫기
```

## 2.2.페이징 처리
- DAUM 뉴스 페이징 처리 : 원하는 keyword를 원하는 페이지만큼 추출

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
driver = webdriver.Chrome()
driver.get('http://daum.net')
# driver.implicitly_wait(0.5) # 0.5초 동안 대기
time.sleep(0.5) # 0.5초 멈춤
driver.find_element(By.CLASS_NAME, 'tf_keyword').click()
query = input('검색할 단어는?')
driver.find_element(By.CSS_SELECTOR, 'input[type="text"]').send_keys(query)
driver.find_element(By.CLASS_NAME, 'btn_ksearch').click()
time.sleep(2)
# 뉴스 탭으로 이동
driver.find_element(By.LINK_TEXT, '뉴스').click()
# 뉴스 제목과 link를 list에 추가
news_list = []
pages = int(input('몇 페이지 크롤링할까?'))
for page in range(1, pages+1):
    strongs = driver.find_elements(By.CSS_SELECTOR, 'div.item-title > strong')
    # print(len(strongs))
    for s in strongs:
        a = s.find_element(By.TAG_NAME, 'a')
        title = a.text
        link = a.get_attribute('href')
        news_list.append([title, link])
    page_nav = driver.find_element(By.CSS_SELECTOR, 'div.inner_paging')
    next_page = page_nav.find_element(By.LINK_TEXT, str(page+1))
    next_page.click()
    time.sleep(2)
driver.close()
import pandas as pd
pd.DataFrame(news_list, columns=['title', 'link'])
```

## 2.3.맞춤법 검사기
- 네이버 맞춤법 검사기 이용

```python
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome()
driver.get('https://naver.com/')
time.sleep(1)
elem = driver.find_element(By.CLASS_NAME, 'search_input')
elem.send_keys('맞춤법 검사기')
elem.send_keys(Keys.RETURN) # form 태그의 submit 클릭 효과
textarea = driver.find_element(By.CLASS_NAME, 'txt_gray')
textarea.clear()
textarea.send_keys('안뇽하세요. 반갑숩니다. 걈사합니다')
btn = driver.find_element(By.CLASS_NAME, 'btn_check')
btn.click()
time.sleep(2)

soup = BeautifulSoup(driver.page_source, 'html.parser')
result = soup.select_one('p._result_text.stand_txt').text
print(result)
driver.close()
```

- 맞춤법 검사 전 .txt 읽어 300자 단위로 잘라 list에 넣어 동적 웹크롤링을 한 후 결과 출력

```python
with open('data/ch14_맞춤법검사전.txt', 'r', encoding='utf-8') as f:
    text = f.read()
ready_text = []
while (len(text) > 300):
    temp = text[:300]
    last_dot_index = temp.rfind('\n')
    ready_text.append(text[0:last_dot_index])
    remaining_text = text[last_dot_index+1 : ]
ready_text.append(remaining_text)
# print(ready_text)
print([len(ready) for ready in ready_text])
for ready in ready_text:
    print(len(ready))
```
