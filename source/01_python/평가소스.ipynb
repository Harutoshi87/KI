# 실행화면 2

```python
df_last = pd.read_csv('C:\KI\downloads\shareData\부동산\주택도시보증공사_전국 신규 민간아파트 분양가격 동향_20241231.csv', encoding='cp949')
df_last.head()
```

```python
import re
df_last['분양가격(제곱미터)'] = pd.to_numeric(df_last['분양가격(제곱미터)'].str.strip().str.replace(',', ''))
df_last['평당분양가격']=df_last['분양가격(제곱미터)']*3.3
df_last.drop('분양가격(제곱미터)', axis=1, inplace=True)
df_last['전용면적'] = df_last['규모구분'].apply(lambda x : re.sub('전용면적|이하| ', '', x).replace('제곱미터', 'm²').replace('초과', '~'))
df_last
```

```python
df_last.drop(['규모구분'], axis=1)
df_last = df_last.loc[:, ['지역명', '전용면적', '연도', '월', '평당분양가격']]
```
